```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.svm import SVC, LinearSVC
from sklearn.neighbors import KNeighborsClassifier

data = pd.read_csv('avocado.csv', index_col = [0])
```

```python
data = data.drop('Date',axis = 1)
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data.region = le.fit_transform(data.region)
data.type = le.fit_transform(data.type)
data = data.astype(int)
data.corr()
```

```python

data =  data.drop('year' , axis = 1)
data =  data.drop('region' , axis =1)
data =  data.drop('XLarge Bags' , axis =1)

data.info()
```

```python
X = data.drop('AveragePrice' ,axis =1)
y = data['AveragePrice']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state = 0)
```

```python
#GNB
gnb = GaussianNB()
y_pred = gnb.fit(X_train, y_train).predict(X_test)
acc_gnb = round(gnb.score(X_train, y_train) * 100, 2)
print("Naïve Bayes accuracy is:", acc_gnb)
```

```python
##SVM
svc = SVC()
svc.fit(X_train, y_train)
Y_pred = svc.predict(X_test)
acc_svc = round(svc.score(X_train, y_train) * 100, 2)
print("svm accuracy is:", acc_svc)
```

```python
##KNN
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train, y_train)
Y_pred = knn.predict(X_test)
acc_knn = round(knn.score(X_train, y_train) * 100, 2)
print("KNN accuracy is:",acc_knn)
```

With n_neaighbors = 3, KNN accuracy is %90.77

As I increased n_neighbors the accuracy would decrease so, I decided to stick with n_neighbors = 3. 



```python
#LinearSVC
LinearSVC = SVC(kernel ='linear', max_iter = 500)
Y_pred = LinearSVC.fit(X_train, y_train).predict(X_test)
acc_LinearSVC = round(LinearSVC.score(X_train, y_train) * 100, 2)
print("Linear SVC accuracy is:", acc_LinearSVC)
```
